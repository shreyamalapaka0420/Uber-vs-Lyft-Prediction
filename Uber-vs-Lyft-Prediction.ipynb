### 2. Data cleaning


#### 2.1 Filling Null values

```python
#integer type columns
dataset = rides_df
mask_int = dataset.dtypes == int
columns_int = dataset.columns[mask_int]
print(columns_int)
```

```python
#float type columns
mask_float = dataset.dtypes == float
columns_float = dataset.columns[mask_float]
print(columns_float)
```

```python
dataset.columns
```

```python
#heatmap to detect frequency of null values
plt.figure(figsize=(25, 10))

sns.heatmap(dataset.isnull(), cbar=False)

plt.show()
```

```python
for column in dataset.columns:
    if column in columns_int or column in columns_float:
        mean_value = dataset[column].mean()
        print('Imputing {} with mean of {}'.format(column, mean_value))
        dataset[column].fillna(value=mean_value, inplace=True)
```

```python
flag = dataset.notnull()
flag
```

```python
#total number of null values in all the columns
dataset.isna().sum()
```

Now, you can notice that all the columns are filled and there are no missing values.

#### 2.2 Outlier Detection

```python
#boxplot to detect outliers
plt.figure(figsize=(15,8))
boxplot = dataset.boxplot(column=['distance', 'price', 'surge_multiplier'])  
```

##### Price column has significant amount of outliers

```python
#function to remove outliers
def outlier_check(df,col):
    Q1 = np.percentile(df[col], 25,
                   interpolation = 'midpoint')
 
    Q3 = np.percentile(df[col], 75,
                   interpolation = 'midpoint')
    IQR = Q3 - Q1
    upper = np.where(df[col] >= (Q3+1.5*IQR))
    lower = np.where(df[col] <= (Q1-1.5*IQR))
    df.drop(upper[0], inplace = True)
    df.drop(lower[0], inplace = True)
```

```python
#removing outliers
outlier_check(dataset,'price')
```

```python
#boxplot of price after removing outliers
plt.figure(figsize=(15,8))
boxplot = dataset.boxplot(column=['price'])
```

### 5. Data Pre-Processing

### 5.1 Feature Engineering

```python
#replacing Lyft with 0 and Uber with 1
dataset['cab_type'] = dataset['cab_type'].replace({'Lyft': 0, 'Uber': 1})
```

```python
categorical_columns = ['name','source','destination','timezone']
```

```python
dataset.iloc[0]
```


### 5.1.1 One Hot Encoding

```python
df= pd.get_dummies(data = dataset,
               columns = categorical_columns,
               drop_first =True,
              dtype='int8')
df=df.drop('id',axis=1)
df=df.drop('product_id',axis=1)
df=df.drop('icon',axis=1)
df=df.drop('short_summary',axis=1)
df=df.drop('long_summary',axis=1)
```

```python
df
```

### 6. Model Building

### 6.1 Train Validation Test Split

```python
from sklearn.model_selection import train_test_split
X=df.drop(['price','datetime'],axis=1)
y=df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = pd.DataFrame(scaler.transform(X_train), columns=X.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X.columns)
```

### 6.2. Linear Regression

```python
#Importing the Linear Regression class
from sklearn.linear_model import LinearRegression
```

```python
#fitting the model to training dataset
linear_reg=LinearRegression()
linear_reg.fit(X_train,y_train)
```

```python
#prediction on test dataset
y_pred=linear_reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print("M score is %f"%mean_squared_error(y_test,y_pred))
print("R2 score is ",r2_score(y_test,y_pred))
MSE_linear=mean_squared_error(y_test,y_pred)
R2_score_linear=r2_score(y_test,y_pred)
```

```python
#cross-validation scores
#using a five_fold cross-validation
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(linear_reg, X_train, y_train, cv=5)
pd.DataFrame(cv_scores,columns=['Score'])
```

```python
#We can observe that cross_validation scores are very close to eachother
```

### 6.3 Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
X_t=X_train[:6000]
y_t=y_train[:6000]

rf_reg = RandomForestRegressor(n_estimators = 500, random_state = 0, n_jobs = -1)
rf_reg.fit(X_t, y_t)
```

```python
y_pred_regfor=rf_reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print("M score is %f"%mean_squared_error(y_test,y_pred_regfor))
print("R2 score is ",r2_score(y_test,y_pred_regfor))
MSE_rf_reg=mean_squared_error(y_test,y_pred_regfor)
R2_score_rf_reg=r2_score(y_test,y_pred_regfor)
```

### 6.4 Support Vector Machine

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
sv_reg = SVR()
```

Defining the parametres for SVR and using GridSearchCV to find the best parametres for our model

```python
params = {'kernel': ('linear', 'rbf','poly'), 'epsilon':[0.1,0.5,0.3]}
clf = GridSearchCV(sv_reg, params)
clf.fit(X_t, y_t)
```

```python
y_pred_sv_reg=clf.predict(X_test)
```

```python
print("M score is %f"%mean_squared_error(y_test,y_pred_sv_reg))
```

```python
print("R2 score is ",r2_score(y_test,y_pred_sv_reg))
```

```python
#Best parametres for our SVR model
clf.best_params_
```

```python
MSE_svmr=mean_squared_error(y_test,y_pred_sv_reg)
R2_score_svm=r2_score(y_test,y_pred_sv_reg)
```

### 6.5 Model Comparison

```python
#preparing dataset for models comparision
labels=["Linear Regression","Random Forest Regressor","Support Vector Regression"]
MSE_scores=[MSE_linear,MSE_rf_reg,MSE_svmr]
R2_scores=[R2_score_linear,R2_score_rf_reg,R2_score_svm]
```

#### 6.5.1 Comparing MSE scores

```python
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(16,8))
fig.tight_layout(pad=10.0)
ax1.bar(labels, MSE_scores,width = 0.4,color=['blue', 'orange','green'])
ax2.pie(MSE_scores,labels=labels)
```

#### 6.5.2 Comparing R2 scores

```python
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(16,8))
fig.tight_layout(pad=10.0)
ax1.bar(labels, R2_scores,width = 0.4,color=['blue', 'orange','green'])
ax2.pie(R2_scores,labels=labels)
```
